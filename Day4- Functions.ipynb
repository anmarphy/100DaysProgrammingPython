### Cuadrado de un número

```python
def cuadrado(x):
    return x**2
```

```python
def parabola(x,a,b,c):
    return a*x**2+b*x+c
```

```python
n=10
a=2
b=0
c=0
l=[]
for i in range(-n,n):
    l.append(parabola(i, a,b,c))
plt.plot(range(-n,n), l)
```

```python
n=10
v_a=[2,3,-1]
v_b=[0,10]
v_c=[0,-200]
l=[]
x=range(-n, n)

for c in v_c:
    for b in v_b:
        for a in v_a:
            l.append([parabola(i, a,b,c) for i in x])
```

```python
for i in range(0,len(l)):
    plt.plot(x,l[i])
```

### Areas

```python
def area_circulo(r):
    ''''Esta funcion retorna el area de un circulo de radio r'''
    return math.pi*r**2
```

```python
radio=0
area=0
l=[]
while area<100:
    l.append(area)
    radio+=1
    area=area_circulo(radio)
```

```python
plt.plot(l)
plt.title('Áreas menores de 100')
plt.show()
```

```python
for i in range(1,5):
    print('El area de un circulo de radio', i, 'es', area_circulo(i))
    i+=1
```

## El condicional If

```python
def valor_absoluto(x):
    if x==0:
        return 0
    elif x<0:
        return -x
    else: 
        return x
```

```python
l=[]
for i in range(-10, 10):
    print(i, valor_absoluto(i))
```

### Estructuras If enlazadas

```python
def precio(unidades, valor_unidad):
    if 10>=unidades>5:
        precio=unidades*0.95*valor_unidad
    elif 20>=unidades>10:
        precio=unidades*0.9*valor_unidad
    elif unidades>20:
        precio=unidades*0.9*valor_unidad
    else: precio=unidades*valor_unidad
    return precio
```

```python
unidades=[1,2,5,6,10,11,21]
for i in unidades:
    print('El precio por ', i, 'unidades es', precio(i, 1000))
```

```python
def area_por_gallina(area, gallinas):
    '''Dado el área del corral de unas gallinas y el número de gallinas en el corral determine
    el área que le corresponde a cada gallina'''
    return area/gallinas
```

```python
 area_por_gallina(100, 2)
```

```python
def max_tres_numeros(a,b,c):
    '''Dados tres números reales calcule el máximo de los tres'''
    if a>b and a>c:
        return a
    if b>a and b>c:
        return b
    else: return c
```

```python
def torque(l_largo, l_corto, peso_largo):
    '''Dadas las longitudes de los dos brazos de una palanca y el peso puesto en el brazo 
    más largo de la palanca, calcular el peso que se puede poner en el brazo más corto para que
    la palanca quede en equilibrio.'''
    return (l_largo*peso_largo)/l_corto
```

```python
torque(100, 20, 100)
```

```python
def long_corto(peso_1, peso_2, longitud_total):
    '''Dados los pesos que se pueden poner en cada uno de los lados de la palanca y la longitud
    total de la palanca determinar la longitud del brazo mas corto para que la
    palanca quede en equilibrio'''
    longitud_corta=peso_2*longitud_total/ (peso_1+peso_2)
    return longitud_corta
```

```python
long_corto(100, 20, 100)
```

```python
def punto_circunferencia(x,y,r, x_check, y_check):
    '''Dado el centro y el radio de un círculo, determinar si un punto pertenece o no al círculo.'''
    if (x-x_check)**2 + (y-y_check)**2< r**2 :
        print('Pertenece a la circunferencia')
    else: print('Este punto cae fuera de la circunferencia')
```

```python
punto_circunferencia(1,1,5, 2, 5)
```

```python
punto_circunferencia(1,1,5, 2, 9)
```
