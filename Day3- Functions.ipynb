```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Una relación R de A en B es un subconjunto del producto cartesiano A x B. En donde A es el conjunto de salida o Dominio y B es el rango. Las propiedades que satisface una relación la caracteriza.
* Reflexiva
* Simétrica
* Antisimétrica 
* Transitiva

Aquellas relaciones que satisfacen que para un elemento del conjunto de salida A le corresponde uno y sólo un elemento en el conjunto de llegada B se denominan Funciones.

* Inyectiva: A un elemento del conjunto B le corresponde una sola preimagen del conjunto A.
* Sobreyectiva: Cada elemento del conjunto B tiene una preimagen del conjunto A.
* Total: Cada elemento del conjunto A tiene una imagen del conjunto B
* Biyectiva: Sii y sólo si, f es inyectiva, sobreyectiva y total.

```python
def plot_identidad(fin):

    def identidad(x):
        return x

    imagen=[]
    for i in range(0, fin):
        imagen.append(identidad(i))
    
    plt.plot(range(0,fin), imagen)
    plt.title('Identidad')
    return 
```

```python
plot_identidad(50)
```

```python
def plot_valor_absoluto(valor):

    def valor_absoluto(numero):
        if numero>0:
            return numero
        else: return -numero
    
    lista = []
    for i in range(-valor, valor):
        lista.append(valor_absoluto(i))    
    
    plt.plot(range(-valor, valor), lista)
    plt.title('Valor absoluto')
    return 
```

```python
plot_valor_absoluto(50)
```

```python
def plot_potencia(base, exponente):
    base=base

    def potencia(base, n):
        return base**n

    lista=[]
    for value in range(-exponente,exponente):
        lista.append(potencia(base, value))
       
    plt.plot(range(-exponente, exponente), lista)
    plt.title('Potencia')
    return 
```

```python
plot_potencia(3, 10)
```

```python
l=[]
n=10
m=1
b=2

for x in range(-n,n):
    l.append(m*x+b)
plt.plot(l, range(-n, n))
```

```python
y=[]
pendiente=2
intercepto=0
n=50

for x in range(-n,n):
    y.append(pendiente*x+intercepto)
plt.plot(range(-n,n),y)
plt.title('Line')
```

```python
x=range(0,12)
slopes=range(0,10)
interceptos=[0,20]
salida=[]
for intercepto in interceptos:
    for slope in slopes:
        salida.append([slope*i+intercepto for i in x])
```

```python
for i in range(0,len(salida)):
    plt.plot(x,salida[i])
```
